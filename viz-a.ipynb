```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
planets = sns.load_dataset('planets')
```

```python
planets
```

```python
sns.relplot(x='year', y='distance', data=planets)
```

```python
sns.relplot(x='year', y='distance', data=planets)
```

```python
combined_years = planets.groupby('year', as_index=False).sum()
combined_years
```

#### relation between the far distance of the discovered  planed from the earth  and the year of discovering it

```python
sns.relplot(x='year', y='distance', data=combined_years)
```

## Train a model using Seaborn

```python
sns.lmplot(x='year', y='distance', data=combined_years)
```

```python
sns.barplot(x='year', y='distance', data=combined_years)
```

#### relation between the number of the discovered planet and the year of discovering it

```python
sns.relplot(x='year', y='number', data=combined_years)
```
