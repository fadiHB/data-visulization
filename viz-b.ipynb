```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
city_temperature = pd.read_csv('city_temperature.csv')
df = city_temperature[['Year','Country','City','AvgTemperature']]
jordan_data = df[df['Country']=='Jordan']
jordan_Amman_data = jordan_data[jordan_data['City']=='Amman']
jordan_Amman_data
```

```python
sns.relplot(x='Year', y='AvgTemperature', data=jordan_Amman_data)
```

```python
combined_years = jordan_Amman_data.groupby('Year', as_index=False).sum()
combined_years
```

```python
sns.relplot(x='Year', y='AvgTemperature', data=combined_years)
```

```python
sns.barplot(x='Year', y='AvgTemperature', data=combined_years)
```
